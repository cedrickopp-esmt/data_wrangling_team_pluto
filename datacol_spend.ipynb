```python
import wbgapi as wb
import pandas as pd
import time

# List of indicators
indicators = [
    "EG.EGY.PRIM.PP.KD", "EG.ELC.ACCS.RU.ZS", "EG.ELC.ACCS.UR.ZS", "EG.ELC.ACCS.ZS",
    "EG.ELC.COAL.ZS", "EG.ELC.FOSL.ZS", "EG.ELC.HYRO.ZS", "EG.ELC.LOSS.ZS",
    "EG.ELC.NGAS.ZS", "EG.ELC.NUCL.ZS", "EG.ELC.PETR.ZS", "EG.ELC.RNEW.ZS",
    "EG.ELC.RNWX.KH", "EG.ELC.RNWX.ZS", "EG.FEC.RNEW.ZS", "EG.GDP.PUSE.KO.PP",
    "EG.GDP.PUSE.KO.PP.KD", "EG.IMP.CONS.ZS", "EG.USE.COMM.CL.ZS", "EG.USE.COMM.FO.ZS",
    "EG.USE.COMM.GD.PP.KD", "EG.USE.CRNW.ZS", "EG.USE.ELEC.KH.PC", "EG.USE.PCAP.KG.OE"
]

# Initialize a list to store all individual DataFrames
all_dfs = []

# Loop through indicators in pairs and create separate DataFrames
for i in range(0, len(indicators), 2):
    pair_index = i // 2 + 1  # Calculate the pair number (1-12)
    batch = indicators[i:i+2]
    print(f"Fetching data for pair {pair_index}: {batch}")

    # Create a separate DataFrame for each pair
    df_name = f"worldb_energy{pair_index}_df"
    globals()[df_name] = wb.data.DataFrame(
        batch,
        time=range(2000, 2020),
        skipBlanks=True,
        labels=True,
        columns='series'
    ).reset_index()

    # Add to the list of all DataFrames
    all_dfs.append(globals()[df_name])

    # Print the shape of the current DataFrame
    print(f"{df_name} shape: {globals()[df_name].shape}")

    # Add a small delay to avoid rate limiting
    time.sleep(1)

# Combine all DataFrames into one
worldb_energy_df = pd.concat(all_dfs, ignore_index=True)
```

```python
import wbgapi as wb
import pandas as pd
import time

# List of environmental finance indicators
envfin_indicators = [
    "NY.GDP.MKTP.CD",
    "NY.GDP.MKTP.CN",
    "NY.GDP.MKTP.KD.ZG",
    "SP.POP.TOTL",
    "NY.ADJ.DCO2.CD",
    "NY.ADJ.DCO2.GN.ZS",
    "NY.ADJ.DPEM.CD",
    "NY.ADJ.DPEM.GN.ZS",
    "DT.NFL.UNEP.CD"
]

# Initialize a list to store all individual DataFrames
all_envfin_dfs = []

# Since we have an odd number of indicators (9), we'll process them in pairs
# and the last one will be processed alone
for i in range(0, len(envfin_indicators), 2):
    pair_index = i // 2 + 1  # Calculate the pair number (1-5)
    batch = envfin_indicators[i:i+2]
    print(f"Fetching data for envfin pair {pair_index}: {batch}")

    # Create a separate DataFrame for each batch
    df_name = f"worldb_envfin{pair_index}_df"

    # Handle the case where we have only one indicator left
    if len(batch) == 1:
        print(f"Fetching single indicator: {batch[0]}")

    globals()[df_name] = wb.data.DataFrame(
        batch,
        time=range(2000, 2020),
        skipBlanks=True,
        labels=True,
        columns='series'
    ).reset_index()

    # Add to the list of all DataFrames
    all_envfin_dfs.append(globals()[df_name])

    # Print the shape of the current DataFrame
    print(f"{df_name} shape: {globals()[df_name].shape}")

    # Add a small delay to avoid rate limiting
    time.sleep(1)

# Combine all DataFrames into one
worldb_envfin_df = pd.concat(all_envfin_dfs, ignore_index=True)

# Display information about the final DataFrame
print("\nFinal combined environmental finance DataFrame:")
print(f"Shape: {worldb_envfin_df.shape}")
print("Columns:", worldb_envfin_df.columns.tolist())
print("\nSample data:")
print(worldb_envfin_df.head())






```
